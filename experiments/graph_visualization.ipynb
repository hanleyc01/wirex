# Visualization of Experimental Data

In this notebook we will be visualizing the experimental data produced by the `experiments` module.
For more information about the module itself, see the documentation contained in each of the implementation
files. The results are serialized in the `/experiments/results` directory.

```python
import json

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
with open("./results/mnisttest.json", "r") as f:
    serialized_results = json.load(f)


data = pd.DataFrame(serialized_results)
data["query"] = data["query"].map(lambda x: np.array(x))
data["result"] = data["result"].map(lambda x: np.array(x))
data.head()
```

```python
plt.title("Similarity as a function of number of entries stored")
plt.plot(data["patterns_stored"], data["cosine_similarity"])
plt.xlabel("# of entries stored")
plt.ylabel("Cosine similarity")
plt.show()
```
